```python
# Coffee dataset

import pandas as pd
import os
import requests
import csv
from bs4 import BeautifulSoup
```

```python

url = 'https://cuela.coffee/collections/coffees-ed-digital?sort_by=title-ascending'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
product_items = soup.findAll('product-item')
names = []
links = []
for product in product_items:
    link = product.find('a')['href']
    name = product.find('img')['alt']
    names.append(name)
    links.append(link)
    #print(product_name, product_link)
df = pd.DataFrame({'Name': names, 'Links': links})
df = df.drop(df.tail(10).index)
df
```

```python
for link in links:
    product_url = 'https://cuela.coffee' + link
    product_page = requests.get(product_url)
    product_soup = BeautifulSoup(product_page.content, 'html.parser')
    
    title_name = product_soup.find('title').text.strip()
    file_name = f'{title_name}.html'
    with open(file_name, 'w') as f:
        f.write(product_page.content.decode())

```

```python
coffee_data = []
for link in links:
    product_url = 'https://cuela.coffee' + link
    product_page = requests.get(product_url)
    product_soup = BeautifulSoup(product_page.content, 'html.parser')
    
    coffee_name = product_soup.find('h1', {'class': 'product-single__title'}).text.strip()
    country = product_soup.find('span', {'class': 'p-country'}).text.strip()
    roast = product_soup.find('span', {'class': 'p-roast'}).text.strip()
    roaster = product_soup.find('span', {'class': 'p-roaster'}).text.strip()
    process = product_soup.find('span', {'class': 'p-process'}).text.strip()
    variety = product_soup.find('span', {'class': 'p-variety'}).text.strip()
    description = product_soup.find('div', {'class': 'p-rte'}).text.strip()
    price = product_soup.find('span', {'class': 'price-item price-item--regular'}).text.strip()
    weight = product_soup.find('span', {'class': 'product-single__weight'}).text.strip()
    
    
    coffee_data.append({
        'name': coffee_name,
        'price': price,
        'weight': weight,
        'variety': variety,
        'country': country,
        'roast': roast,
        'roaster': roaster,
        'process': process,
        'description': description,
        'url': product_url
        
    })

coffee_df = pd.DataFrame(coffee_data)
print(coffee_df)
```
