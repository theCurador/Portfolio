```python
# Process matches in increments of 50
for i in range(0, len(match_ids), 50):
    # Get a batch of 50 match IDs
    batch = match_ids[i:i+50]
    
    # Loop through each match ID in the batch
    for match_id in batch:
        url = base_url.format(match_id)
        response = requests.get(url)
        
        # Check if the response was successful
        if response.status_code == 200:
            try:
                data = response.json()
                file_name = f"{match_id}.json"
                with open(file_name, "w") as f:
                    json.dump(data, f)
            except json.JSONDecodeError:
                print(f"Error parsing JSON for match ID: {match_id}")
        else:
            print(f"Error fetching match ID: {match_id}")
```
