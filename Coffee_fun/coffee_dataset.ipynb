```python

url = 'https://cuela.coffee/collections/coffees-ed-digital?sort_by=title-ascending'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
product_items = soup.findAll('product-item')
names = []
links = []
for product in product_items:
    link = product.find('a')['href']
    name = product.find('img')['alt']
    names.append(name)
    links.append(link)
    #print(product_name, product_link)
df = pd.DataFrame({'Name': names, 'Links': links})
df = df.drop(df.tail(10).index)
df
```
