### Interpreting Data Description 

The results show 8 numbers for each column in your original dataset. The first number, the **count**, shows how many rows have non-missinng values. 

Missing values arise for many reasons. For example, the size of the 2nd bedroom wouldn't be collected when surveying a 1 bedroom house. We'll come back to the topic of missing data. 

The second value is the **mean**, which is the average. Under that, **std** is the standard deviation, which measures how numerically spread out the values are. 

To interpret the **min, 25%, 50%, 75%** and **max** values, imagine sorting each column from lowest to highest value. The first(smallest) value is the min. If you go a quarter way through the list, you'll find a number that is bigger than 25% of the values and smaller that 75% of the values. That is the 25% values (pronounced "25th percentile"). The 50th and the 75th percentiles are defined analogously, adn the **max** is the largest number.

To calculate the avg lot size rounded to the nearest int, we can use the 'round()' function in Python. 

**avg_lot_size**

This code assumes that 'home_data' is a DataFrame containing a column named 'LotArea' that represents the lot size of each home. The '.mean()' method calculates the average lot size, and the 'round()' function rounds the result to the nearest integer. The resulting 'avg_lot_size' variable would contain the average lot size rounded to the nearest integer.

To calculate the age of the newest home, we would need to know the year in which it was built. If we assume that the newest home was built in the current year(2023), then its age would be zero. 

**new_home_age** variable would assume that the 'YearBuilt' column in 'home_data' contains the year inwhich each home was built, and that the 'max()' method returns the yar of the newest hme in the dataset. The resulting 'new_home_age' variable would contain the age of the newest home in years. 


```python
# What is the average lot size (rounded to neares integer)?

avg_lot_size = round(home_data['LotArea'].mean())

# As of today, how old is the newest home(current year - the date in which it was built)
new_home_age = 2023 - home_data['YearBuilt'].max()


print("The average lot size is:",avg_lot_size)
print("The age of the newest home is:",new_home_age)
```
