To create a simulation, we will need to define some variables that will represent the peak performance of each player. These variables will be used to generate random numbers that will simulate the performance of each player in a one-on-one game.

For example, we can define the following variables:

MJ_peak_score: represents the highest score Michael Jordan has ever achieved in a game
LJ_peak_score: represents the highest score LeBron James has ever achieved in a game
Barea_peak_score: represents the highest score JJ Barea has ever achieved in a game
Arroyo_peak_score: represents the highest score Carlos Arroyo has ever achieved in a game
We can use these variables to generate random scores for each player in a one-on-one simulation. We will also need to define the number of simulations to run, and keep track of the number of times each player wins.

Let's start by defining these variables and the number of simulations to run:

python


```python
import random

# Define peak scores for each player
MJ_peak_score = 69
LJ_peak_score = 61
Barea_peak_score = 30
Arroyo_peak_score = 28

# Define number of simulations to run
num_simulations = 10000

# Initialize counters for each player's wins
MJ_wins = 0
LJ_wins = 0
Barea_wins = 0
Arroyo_wins = 0
```

Now, let's create a function that will simulate a one-on-one game between two players. The function will take as input the two players, generate random scores for each player based on their peak scores, and return the winner of the game.

```python
def simulate_one_on_one(player1, player2):
    # Generate random scores for each player based on their peak scores
    player1_score = random.randint(0, player1+'_peak_score')
    player2_score = random.randint(0, player2+'_peak_score')
    
    # Determine the winner of the game based on the scores
    if player1_score > player2_score:
        return player1
    else:
        return player2
```

Finally, let's run the simulations and keep track of the number of times each player wins:

```python
# Run simulations
for i in range(num_simulations):
    # Simulate one-on-one game between two random players
    players = random.sample(sample_player_ids, 2)
    winner = simulate_one_on_one(players[0], players[1])
    
```
