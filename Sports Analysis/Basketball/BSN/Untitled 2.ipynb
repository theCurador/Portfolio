```python
import requests
from bs4 import BeautifulSoup

def scrape_play_by_play(url):
    # Make a request to the webpage
    response = requests.get(url)

    # Parse the HTML content using Beautiful Soup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the element that contains the play-by-play data
    play_by_play_div = soup.find('div', {'class': 'play_by_play'})

    # Extract the text content of the element
    play_by_play_text = play_by_play_div.get_text()

    # Split the text into individual plays
    plays = play_by_play_text.split('\n\n')

    # Return the list of plays
    return plays
```

```python
url = 'https://fibalivestats.dcd.shared.geniussports.com/webcast/BSN/2247780'
plays = scrape_play_by_play(url) 
print(plays)
```
